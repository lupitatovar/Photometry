```python
#HR Diagram of Sun Like Stars that were chosen
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
from pylab import *


teff,logg = np.loadtxt('randomstarset.txt',skiprows=1,usecols=(0,1),unpack=True)
plt.figure(figsize=(8,5))

plt.gcf().subplots_adjust(bottom=0.17, wspace=0.0, top=0.86, right=0.94, left=0.16)
ax = plt.gca()

ax.set_xlim(12000, 2000)
ax.set_ylim(6,-1)
ax.set_ylabel('Surface Gravity (log g)', fontsize=18)
ax.set_xlabel('Temperature (K)', fontsize=18)
#ax.set_title ('Kepler Stars',fontsize=22)

plt.rcParams['axes.linewidth']=5
plt.rcParams['xtick.major.width'] = 2
plt.rcParams['ytick.major.width'] = 2
plt.rcParams['xtick.minor.width'] = 2
plt.rcParams['ytick.minor.width'] = 2
plt.rc('xtick.major', size=8, pad=8)
plt.rc('xtick.minor', size=6, pad=5)
plt.rc('ytick.major', size=8, pad=8)
plt.rc('ytick.minor', size=6, pad=5)


plt.plot(teff,logg,'o',markersize=2,color='darkblue',alpha=0.1,markeredgecolor='darkblue')
plt.semilogx()

ax.set_xticks([12000,11000,10000,9000,8000,7000,6000,5000,4000,3000,2000])
ax.set_xticklabels(['12000','','','9000','','','6000','','4000','3000','2000'], fontsize=16)
ax.set_yticks(np.arange(-1, 6+1, 1))
ax.set_yticklabels(np.arange(-1, 6+1, 1), fontsize=16)

plt.fill([5628,5928,5928,5628,5628], [5.9,5.9,4.2,4.2,5.9],'y',zorder=3, alpha=0.5)
#rect = plt.Rectangle((5628, 4.1),150,3, color='red')
#plt.gca().add_patch(rect)

#add orange dot where the sun is!
ax.plot((5777), (4.438), 'o', color='orange')


plt.show()

```

```python
#Histograms of Temperature, Surface Gravity, Magnitude, Metallicity
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

teff,logg,mag,metal = np.loadtxt('sunstarset.txt',delimiter=',',usecols=(1,2,3,4),skiprows=(1),unpack=True)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8,5))
ax0, ax1, ax2, ax3 = axes.flat

#Temperature Histogram

tempbins=np.arange(5628,5928,25)

ax0.hist(teff, tempbins, histtype='bar',fill=False) 
ax0.set_title('Temperature')
ax0.axvline(5777, color='black', linestyle='dashed', linewidth=2)

#Surface Gravity Histogram

logbins=np.arange(4,5.2,0.1)

ax1.hist(logg, logbins, histtype='bar',fill=False)
ax1.set_title('Surface Gravity')
ax1.axvline(4.438, color='black', linestyle='dashed', linewidth=2)

#Magnitudes Histogram

magbins=np.arange(12,17.1,0.5)

ax2.hist(mag,magbins,histtype='bar',fill=False)
ax2.set_title('Magnitude')

# Metallicity Histogram

metalbins=np.arange(-2,1,0.2)

ax3.hist(metal, metalbins, histtype='bar',fill=False)
ax3.set_title('Metallicity')
ax3.axvline(0.0134, color='black', linestyle='dashed', linewidth=2.5)

plt.tight_layout()
plt.show()

#plt.savefig
#put into sharelatex
```
